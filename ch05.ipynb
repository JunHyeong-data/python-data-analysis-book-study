- loc: 레이블 기반 인덱싱

  - 인덱스와 컬럼의 이름을 사용하여 데이터를 선택합니다.

  - 슬라이싱 시 끝 값도 포함됩니다.

  - 예시: df.loc[2, 'column_name']은 인덱스가 2인 행과 'column_name' 열을 선택합니다.

- iloc: 위치 기반 인덱싱

  - 행과 열의 정수 위치(0부터 시작하는 인덱스)를 사용하여 데이터를 선택합니다.

  - 슬라이싱 시 끝 값은 포함되지 않습니다.

  - 예시: df.iloc[2, 3]은 3번째 행과 4번째 열의 값을 선택합니다.

### 산술 연산과 데이터 정렬
- 판다스는 서로 다른 색인을 가지고 있는 객체 간의 산술 연산을 간단하게 처리

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1],
               index=["a", "c", "e", "f", "g"])
print(s1)
s2
```

```python
s1 + s2
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list("bcd"),
                   index=["Ohio", "Texas", "Colorado"])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list("bde"),
                   index=["Utah", "Ohio", "Texas", "Oregon"])
df1
```

```python
df2
```

```python
df1 + df2
# 두 DataFrame을 더하면 각 DataFrame에 있는 색인과 열이 하나로 합쳐진다.
```

```python
df1 = pd.DataFrame({"A": [1, 2]})
df2 = pd.DataFrame({"B": [3, 4]})
df1
df2
df1 + df2
# 공통 열이나 행 레이블이 없는 DataFrame을 더하면 결과는 아무것도 나타나지 않는다.
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)),
                   columns=list("abcd"))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                   columns=list("abcde"))
df2.loc[1, "b"] = np.nan
print(df1)
df2
```

```python
df1 + df2
# 두 결과를 더하면 겹치지 않는 부분은 NaN이다
```

```python
df1.add(df2, fill_value=0)
# 연산에서 누락된 값은 fill_value 값으로 대체된다.
```

```python
print(1 / df1)
df1.rdiv(1)
```

```python
df1.reindex(columns=df2.columns, fill_value=0)
```

### 함수 적용과 매핑

```python
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
print(frame)
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()

frame.apply(f1)
```

```python
frame.apply(f1, axis="columns")
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=["min", "max"])
frame.apply(f2)
```

```python
def my_format(x):
    return f"{x:.2f}"

frame.applymap(my_format)
```

```python
frame["e"].map(my_format)
```
