### 표시자, 더미 변수 계산

get_dummies는 주로 Pandas 라이브러리에서 사용되며, 범주형(categorical) 데이터를 머신러닝 모델 등이 이해할 수 있도록 숫자형의 가변수(dummy variables) 또는 원-핫 인코딩(one-hot encoded) 형태로 변환하는 함수입니다.

핵심 기능:

- 원래 데이터의 각 범주(예: '색상' 열의 '빨강', '파랑')를 새로운 열로 만듭니다.
- 각 행에서 해당 범주에 속하면 1, 그렇지 않으면 0으로 표시합니다.
주요 목적:

- 문자열과 같은 범주형 데이터를 숫자 데이터로 변환하여 통계 모델이나 머신러닝 알고리즘에 사용할 수 있도록 합니다.
- 범주 간의 순서가 없는 데이터를 처리할 때, 모델이 잘못된 순서 정보를 학습하는 것을 방지합니다.

```python
df = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                   "data1": range(6)})
print(df)
pd.get_dummies(df["key"], dtype=float)
```

```python
dummies = pd.get_dummies(df["key"], prefix="key", dtype=float)
df_with_dummy = df[["data1"]].join(dummies)
df_with_dummy
```

### 정규 표현식
- 정규 표현식은 텍스트에서 문자열 패턴을 찾는 유연한 방법을 제공
- regex라 부르는 단일 표현식은 정규 표현식 언어로 구성된 문자열
- re 모듈 함수는 패턴 매칭, 치환, 분리 세 가지로 나눌 수 있다


```python
import re
text = "foo    bar\t baz  \tqux"
re.split(r"\s+", text)
# 여러 가지 공백 문자(탭, 스페이스, 개행 문자)가 포함된 문자열을 나누고 싶다면
# 하나 이상의 공백 문자를 의미하는 \s+를 사용해서 문자열을 분리한다
```

```python
regex = re.compile(r"\s+")
regex.split(text)
```

```python
regex.findall(text)
# 정규 표현식에 매칭되는 모든 패턴의 목록을 얻고 싶다면 findall 메서드를 사용
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""
pattern = r"[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}"

# re.IGNORECASE makes the regex case insensitive
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
regex.findall(text)
```

```python
m = regex.search(text)
print(m)
text[m.start():m.end()]
```

```python
print(regex.match(text))
```

```python
print(regex.sub("REDACTED", text))
```

```python
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
m = regex.match("wesm@bright.net")
m.groups()
```

```python
regex.findall(text)
```

```python
print(regex.sub(r"Username: \1, Domain: \2, Suffix: \3", text))
```
